# >>> Data Prep (Splitting X & y, train & test)

```python
# check value balance of y
y = "Expensive"  # -----------------> INPUT of intended y <-----------------------------------< INPUT y
# balance = data[y].value_counts()
# balance
percentage = data[y].value_counts(normalize=True) * 100
percentage
```

```python
# delete data with NaNs in y feature
data.dropna(subset=[y], inplace=True)

# define y (and X)
y = data.pop(y)
X = data

# feature engineering
# X.loc[:, "Cabin"] = X.Cabin.str[0]  # extracting only the first character of each string
# X.loc[X.loc[:, "Heating"] == "Floor", "Heating"] = "NA"
X['Street'] = X['Street'].replace({'Pave': 1, 'Grvl': 0})
X['CentralAir'] = X['CentralAir'].replace({'Y': 1, 'N': 0})

# ---> general INPUTS and for data split <--------------------------------------------------< INPUTS split
seed = 123        # random_state
split = 0.2       # split portion for test_dataset

# split data to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split, random_state=seed)
```

```python
# INPUTS for cross validation and quality check/score <---------------------------------------< INPUTS folds & model_score
cv_folds = 10
score_metric = "accuracy"  # ["precision", "recall", "f1", "roc_auc"]
model_scores_dict = {}
```

# Preprocessor (upper reach)

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
# define numerical and categorial branch
num_features = data.select_dtypes(include="number").columns
cat_features = data.select_dtypes(include="object").columns

# define transformers for each branch
num_branch = make_pipeline(SimpleImputer(), StandardScaler())
cat_branch = make_pipeline(SimpleImputer(strategy="constant", fill_value="NA"), OneHotEncoder(handle_unknown="ignore", sparse_output=False)).set_output(transform='pandas')

# create branches for preprocessor
preprocessor = make_column_transformer(
                    (num_branch, num_features),
                    (cat_branch, cat_features))
```

# Estimators (downstream)

## Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

pipe = make_pipeline(preprocessor, SVC(class_weight="balanced"))
grid = {
    "columntransformer__pipeline-1__simpleimputer__strategy":["mean"],
    'svc__C': [10],               # svc__C': [0.5, 1, 10, 100, 1000], # C values must be > 0
    'svc__kernel': ['rbf'],       # 'svc__kernel': ['linear', 'rbf', 'poly'],
    'svc__gamma': [0.0002],       # 'svc__gamma': ['scale', 'auto', 0.0001, 0.001, 0.01, 0.1, 1, 10, 100] samll gamma = more generalized decision boundary
}

# cross validation
SVC_model = GridSearchCV(pipe, grid, cv=cv_folds, scoring=score_metric, verbose=1)
SVC_model.fit(X_train, y_train)

# scores dictionary
model_scores_dict["SVC"] = round(SVC_model.best_score_, 4)
```

```python
# SVC_model.best_params_
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

pipe = make_pipeline(preprocessor, RandomForestClassifier())
grid = {
    "columntransformer__pipeline-1__simpleimputer__strategy":["mean"],
    'randomforestclassifier__max_depth': [8],                    # levels of trees
    'randomforestclassifier__min_samples_leaf': [5],             # min nr of obs for a leafe    
    'randomforestclassifier__min_samples_split': [10],           # min nr of obs for a node to be split again
    'randomforestclassifier__n_estimators': [34]                 # number of trees in the forest
}

# cross validation
forest_model = GridSearchCV(pipe, grid, cv=cv_folds, scoring=score_metric, verbose=1)
forest_model.fit(X_train, y_train)

# scores dictionary
model_scores_dict["forest"] = round(forest_model.best_score_, 4)
```

## Train-Test-Accuracy

```python
models = [SVC_model, forest_model]
model_names = ["SVC_model", "forest_model"]
```

```python
from sklearn.metrics import accuracy_score, f1_score, cohen_kappa_score , precision_score, recall_score, roc_auc_score #, make_scorer

eval_dict = {}
for m, n in zip(models, model_names):
    
    # accuracy
    y_train_pred = m.predict(X_train)
    train_acc = accuracy_score(y_train, y_train_pred)
    y_test_pred = m.predict(X_test)
    test_acc = accuracy_score(y_test, y_test_pred)
    acc_delta = train_acc - test_acc

    # Storing results in eval_dict
    eval_dict[n] = {
        "Training Accuracy": train_acc,
        "Testing Accuracy": test_acc,
        "Accuracy_Delta": acc_delta,
        }
    
eval_df = pd.DataFrame(eval_dict)
eval_df
```

# Evaluators (Test Data)

## <<< Scores

```python
from sklearn.metrics import f1_score, precision_score, recall_score, roc_auc_score, average_precision_score

eval_dict = {}
for m, n in zip(models, model_names):
    y_test_pred = m.predict(X_test)  # prediction

    # precision
    precision = round(precision_score(y_test, y_test_pred), 4)

    # recall
    recall = round(recall_score(y_test, y_test_pred), 4)

    # f1-score
    # f1 = round(f1_score(y_test, y_test_pred), 4)
    f1_micro = round(f1_score(y_test, y_test_pred, average='micro'), 4)
    f1_macro = round(f1_score(y_test, y_test_pred, average='macro'), 4)

    # Storing results in eval_dict
    eval_dict[n] = {
        "Precision": precision,
        "Recall": recall,
        # "F-Score": f1,
        "F1 Micro": f1_micro,
        "F1 Macro": f1_macro,
    }

eval_df = pd.DataFrame(eval_dict)

# metric comment
comments_dict = {
        "Precision": "is about avoiding false positives (accuracy of positives): High precision ‚Üí Low false positive rate",
        "Recall": "is about capturing ALL positives: High recall ‚Üí Low false negative rate",
        # "F-Score": "Harmonic mean of precision and recall for situations where the cost of false positives and false negatives is similar",
        "F1 Micro": " to consider imbalances in class sizes. May be influenced more by the majority class in highly imbalanced datasets",
        "F1 Macro": "provides equal weight to each class, regardless of class size. Useful when to avoid class size biases. Can be sensitive to small classes",
        }
eval_df['metric comment'] = eval_df.index.map(comments_dict)
pd.set_option('display.max_colwidth', 200)  # set column width for comment

eval_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame(eval_dict).T
plt.figure(figsize=(5, 3))
heatmap = sns.heatmap(df.T, annot=True, cmap="Spectral", fmt=".4f", linewidths=0, cbar_kws={"label": "Metric Value"})
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title('Estimator Comparison')
plt.show()
```
